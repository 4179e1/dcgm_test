```python
import pandas as pd
import plotly.graph_objects as go
import sys

def myplot(filename):
    # Load the CSV file
    data = pd.read_csv(filename)

    # Convert TimeStampUs to a human-readable datetime format
    data['TimeStamp'] = pd.to_datetime(data['TimeStampUs'], unit='us')

    # Get unique GPU IDs
    gpu_ids = data['GPUId'].unique()

    # Create figure
    fig = go.Figure()

    # Plot the data for each GPU
    for gpu_id in gpu_ids:
        gpu_data = data[data['GPUId'] == gpu_id].reset_index(drop=True)  # Reindex each gpu_data starting from 0
        print (f"GPU{gpu_id} samples: {len(gpu_data)}")
        fig.add_trace(go.Scatter(x=gpu_data['TimeStamp'], y=gpu_data['Watts'], mode='lines', name=f'GPU {gpu_id}'))

    # Customize layout
    fig.update_layout(
        title='Power Consumption Over Time for Each GPU',
        xaxis_title='Timestamp',
        yaxis_title='Watts',
        xaxis_tickformat='%H:%M:%S.%f',  # Format x-axis tick labels
        hovermode='x',  # Show hover information only for x-axis
        template='plotly_white'  # Set plot style
    )

    # Show the plot
    fig.show()
```

```python
myplot("result.csv")
```
